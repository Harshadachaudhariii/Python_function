**Q. 1  Write a Python function that takes a list of numbers as input and returns the sum of all even numbers in
the list.**

```python
def sum_of_all_even(lst):
  '''
  Calculating the sum of all even number in the input list

  Parameter:
  lst (list): list of inegter number

  Return:
  int : Returns the sums of all even number from the input list
  '''

  sums = 0
  for i in lst:
    if i%2==0:
      sums +=i
  return sums

lst = list(map(int,input("Enter the number separate by space : ").split()))
print("Sums of all even number from list is : ",sum_of_all_even(lst))
```

**Q. 2  Create a Python function that accepts a string and returns the reverse of that string.**

```python
def reverse_string(strs):
  '''
  Reversing string in a given input string

  Parameter:
  strs (string): The input string to be reversed.

  Return:
  str: The reversed string
  '''
  return strs[::-1]

strs = input("Enter the string")
print(reverse_string(strs))
```

**Q. 3  Implement a Python function that takes a list of integers and returns a new list containing the squares of
each number.**

```python
def square_of_each(lst):
  '''
  Calculating square of each number from input list and returning new list

  Parameter:
  lst (list): A list of integer number

  Return:
  list : Return new list with containing the square of each number
  '''
  new_lst = []

  for i in lst:
      new_lst.append(i**2)
  return new_lst

lst = list(map(int, input("Enter the number separate by space : ").split()))
print(square_of_each(lst))
```

**Q. 4 Write a Python function that checks if a given number is prime or not from 1 to 200.**

```python
def is_prime(n):
  '''
  Checking given number is prime or not

  Parameter:
  n (int): the input as an integer number

  Return:
  bool: True if the number is prime, False otherwise.
  '''

  if n <=1:    #if n is less than equal to 1 then not prime number
    return False
  elif n==2:  # if n is 2 return true because 2 is even-prime number
    return True
  else:
    for i in range(2,n):    # loop start from 2 to n-1 checking if n not divide to n-1 number then return false and composite number there then not prime
      if n%i==0:
        return False
  return True   # after dividing number not found divide any number then is prime number

num = int(input("Enter the number "))
if is_prime(num):
  print("Given number is prime number ")
else:
  print("Give number is not prime number or may be composite number")
```

**Q. 5  Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of
terms.**

```python
def fibonacci_series(n):
  '''
  Generating Fibonacci sequence up to a specified number of terms

  Parameter:
  n (int): input of integer number

  Return:
  int: Here yield keyword used to generate output one by one and return one value at once
  '''

  prev_num = 0
  new_num = 1

  for i in range(n):
    yield prev_num
    prev_num, new_num = new_num, prev_num+new_num   # swaping two variable and 2nd variable addition of previous two number

num= int(input("Enter the number "))
fib_series = fibonacci_series(num)
next(fib_series)
```

**Q. 6 Write a generator function in Python that yields the powers of 2 up to a given exponent.**

```python
def power_of_2(exps):
  '''
  Generating powers of 2 up to a given exponent

  Parameter:
  exps (int): The input is a integer number

  Return:
  int: Here yield keyword used to generate output one by one and return one value at once
  '''
  for i in range(exps+1):
    yield 2**i
num = int(input("Enter the number "))
power = power_of_2(num)
next(power)
```

**Q. 7 Implement a generator function that reads a file line by line and yields each line as a string.**

```python
def read_file(file_path):
  '''
  Generating function that reads a file line by line

  Parameter:
  file_path: Input as the file name/path

  Return:
  String: Here yield keyword used to generate output one by one and return one value at once
  These generator function reads line by line return one string at once
  '''
  with open(file_path, 'r') as file:
    for line in file:
      yield line.rstrip('\n')
files = read_file("/content/read.txt")
next(files)
```

**Q. 9  Write a Python program that uses `map()` to convert a list of temperatures from Celsius to Fahrenheit.**

```python
def temp_conversion(celsius):
  '''
  Converting list of temperature from celsius to fahrenheit
  formula: fahren = (celsius *9/5)+32

  Parameter:
  celsius (float): Input of floating number

  Return:
  The return float value converting to celsius to fahrenheit
 '''
  fahren_temp =( celsius*9/5)+32
  return fahren_temp

celsius_temp = list(map(float, input("Enter the temperatue separate by space ").split()))
result = list(map(temp_conversion,celsius_temp))
print("Temperature from celsius to Fahrenheit is : ",result)
```

**Q. 10 Create a Python program that uses `filter()` to remove all the vowels from a given string.**

```python
def remove_vowels(s):
  '''
  Removing all the vowels from a given strings

  Paramater:
  s (str): Input as the string

  Return:
  str: returning string but after removing vowels from the string
  '''
  vowels = ['a' , 'i' , 'o' , 'u' , 'e']
  return ''.join(filter(lambda ch:  ch not in vowels,s))

strs = input("Enter the String ").lower()
results = remove_vowels(strs)
print("Removing all the vowels from a give string ",results)
```

**Q. 11  Imagine an accounting routine used in a book shop. It works on a list with sublists, which look like this:**

**Write a Python program, which returns a list with 2-tuples. Each tuple consists of the order number and the
product of the price per item and the quantity. The product should be increased by 10,-  if the value of the
order is smaller than 100.00 .**

**Write a Python program using lambda and map.**

```python
def accounting(orders):
  '''
  Calculates the total price for each book order.

    Each order is a list containing:
    [order_number, book_title, quantity, price_per_unit]

    If the total price (quantity  price) is less than or equal to 100.00,
    an additional handling fee of 10 is added.

    Parameters:
    orders (list): A list of book orders, where each order is a list.

    Returns:
    list: A list of tuples containing (order number, final price)

  '''
  results = []
  for order in orders:
    order_num = order[0]
    quantity = order[2]
    price = order[3]
    total = quantity*price
    if total <=100.00:
      total +=10
    results.append((order_num, round(total,2)))
  return results

orders =[
    [34587, "Learning Python, Mark Lutz", 4 , 40.95],
    [98762, "Programming Python, Mark Lutz", 5 , 56.80],
    [77226, "Head First Python, Paul Barry", 3, 32.95],
    [88112, "Einfuhrung in Python3, Bernd Klein", 3 , 24.99]
]
result = accounting(orders)
print(result)
```
